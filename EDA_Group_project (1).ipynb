# Exploratory Data Analysis (EDA) – Loan Sanction Dataset

### INTRODUCTION

##### This project performs a complete Exploratory Data Analysis (EDA) on the Loan Sanction Training Dataset. The objective is to understand the data structure, clean the dataset, analyze variable distributions, explore relationships between variables, and derive meaningful insights related to loan approval status.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
categorical_cols = ['Gender', 'Married', 'Dependents', 'Self_Employed', 'Credit_History']

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

numeric_cols = ['LoanAmount', 'Loan_Amount_Term']

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

print(df.isnull().sum())
```

### Observation:

Mode is used for categorical columns.

Median is used for numerical columns to reduce the impact of outliers.

Dataset is now free from missing values.

# Bivariate Analysis 

#### Categorical vs Categorical Analysis

```python
#Loan_Status vs Gender 
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Loan_Status', data=df)
plt.title('Loan Status vs Gender')
plt.show()
```

## Insight:
Male applicants have more approvals, largely due to higher application count.

```python
#Loan_Status Proportion 
plt.figure(figsize=(5,5))
df['Loan_Status'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', startangle=90
)
plt.title('Loan Status Proportion')
plt.ylabel('')
plt.show()
```

## Insight:
Confirms class imbalance with more approved loans.

```python
#Loan_Status vs Credit_History 
pd.crosstab(df['Credit_History'], df['Loan_Status']).plot(
    kind='bar', stacked=True, figsize=(6,4)
)
plt.title('Loan Status vs Credit History')
plt.xlabel('Credit History')
plt.ylabel('Count')
plt.show()
```

## Insight:
Applicants with good credit history have a much higher approval rate.
This is the strongest predictor in the dataset.

### Categorical vs Numerical

```python
#Loan_Status vs ApplicantIncome 
plt.figure(figsize=(6,4))
sns.boxplot(x='Loan_Status', y='ApplicantIncome', data=df)
plt.title('Applicant Income vs Loan Status')
plt.show()
```

## Insight:
Approved applicants tend to have slightly higher income, but overlap exists.

### Loan_Status vs LoanAmount 

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Loan_Status', y='LoanAmount', data=df)
plt.title('Loan Amount vs Loan Status')
plt.show()
```

### Insight:
Moderate loan amounts are more likely to be approved.

## Applicant Income by Property Area

```python
plt.figure(figsize=(6,4))
df.groupby('Property_Area')['ApplicantIncome'].median().plot(kind='bar')
plt.title('Median Applicant Income by Property Area')
plt.xlabel('Property Area')
plt.ylabel('Median Applicant Income')
plt.show()
```

### Insight:
The median applicant income varies across property areas. Rural applicants show a slightly higher median income compared to urban and semiurban applicants. Since the median is less affected by extreme values, this provides a more robust comparison of income levels across property areas.

### MULTIVARIATE ANALYSIS 

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']]
    .corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title('Correlation Heatmap')
plt.show()
```

## Insight
	•	ApplicantIncome and LoanAmount show moderate positive correlation
	•	CoapplicantIncome has weak correlation with LoanAmount
	•	Credit_History shows very low correlation with income variables
	•	No strong multicollinearity is observed

## Scatter Plot: ApplicantIncome vs LoanAmount

```python
plt.figure(figsize=(6,4))
sns.scatterplot(
    x='ApplicantIncome',
    y='LoanAmount',
    hue='Loan_Status',
    data=df
)
plt.title('Applicant Income vs Loan Amount by Loan Status')
plt.show()
```

## Insight
	•	Higher income applicants generally apply for higher loan amounts
	•	Approved and rejected loans overlap significantly
	•	Income alone does not determine loan approval

##  Overall Bivariate Conclusion
	•	Loan approval is influenced by multiple interacting factors, rather than a single variable.
	•	Credit history emerges as the most influential bivariate factor affecting loan approval.
	•	Income and loan amount provide useful context but cannot independently predict loan sanction.
	•	The observed overlaps across groups highlight the importance of considering multiple features jointly.
